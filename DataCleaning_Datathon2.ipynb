# Cleaning the name.

```python
import re

# Function to clean the name column and extract the correct name, bedrooms, beds, baths, and rating
def clean_listing_name(name):
    # Step 1: Extract the first part of the name (before the first "·")
    actual_name = name.split('·')[0].strip()

    # Step 2: Remove the star symbol but keep the rating
    rating_match = re.search(r'★(\d\.\d{2})', name)  # Match star rating like ★4.83
    if rating_match:
        rating_in_name = float(rating_match.group(1))
    else:
        rating_in_name = 2.5  # Default to 2.5 if no rating is found

    # Step 3: Handle "studio" as 1 bedroom, otherwise extract the number of bedrooms
    if 'studio' in name.lower():
        num_bedrooms = 1  # Studio is assumed to be 1 bedroom
    else:
        # Extract number of bedrooms if available
        match_bedrooms = re.search(r'(\d+) bedroom', name)
        num_bedrooms = int(match_bedrooms.group(1)) if match_bedrooms else None

    # Step 4: Extract number of beds
    match_beds = re.search(r'(\d+) beds?', name)
    num_beds = int(match_beds.group(1)) if match_beds else None

    # Step 5: Extract number of baths, including floating-point values (e.g., 1.5 baths)
    match_baths = re.search(r'(\d+(\.\d+)?) baths?', name)  # This will capture numbers like 1 or 1.5
    num_baths = float(match_baths.group(1)) if match_baths else None

    return actual_name, num_bedrooms, num_beds, num_baths, rating_in_name

# Apply the cleaning function to the datasets
vancouver_df['name'], vancouver_df['num_bedrooms'], vancouver_df['num_beds'], vancouver_df['num_baths'], vancouver_df['rating_in_the_name'] = zip(*vancouver_df['name'].apply(clean_listing_name))
london_df['name'], london_df['num_bedrooms'], london_df['num_beds'], london_df['num_baths'], london_df['rating_in_the_name'] = zip(*london_df['name'].apply(clean_listing_name))
bangkok_df['name'], bangkok_df['num_bedrooms'], bangkok_df['num_beds'], bangkok_df['num_baths'], bangkok_df['rating_in_the_name'] = zip(*bangkok_df['name'].apply(clean_listing_name))

# Verify the cleaned data while keeping the first part of the 'name' column
print(vancouver_df[['name', 'num_bedrooms', 'num_beds', 'num_baths', 'rating_in_the_name']].head())
print(london_df[['name', 'num_bedrooms', 'num_beds', 'num_baths', 'rating_in_the_name']].head())
print(bangkok_df[['name', 'num_bedrooms', 'num_beds', 'num_baths', 'rating_in_the_name']].head())
```

# Dropping Unnecessary Columns, Removing Rows with Missing last_review and reviews_per_month, and Filling Missing host_name with "Unknown"

```python
# Check and drop columns if they exist
columns_to_drop = ['neighbourhood_group', 'license']

# For Vancouver
vancouver_df = vancouver_df.drop(columns=[col for col in columns_to_drop if col in vancouver_df.columns], axis=1)
# For London
london_df = london_df.drop(columns=[col for col in columns_to_drop if col in london_df.columns], axis=1)
# For Bangkok
bangkok_df = bangkok_df.drop(columns=[col for col in columns_to_drop if col in bangkok_df.columns], axis=1)


#drop rows where both 'last review' and 'review_per_month' are missing
vancouver_df = vancouver_df.dropna(subset=['last_review', 'reviews_per_month'])
london_df = london_df.dropna(subset=['last_review', 'reviews_per_month'])
bangkok_df = bangkok_df.dropna(subset=['last_review', 'reviews_per_month'])

#filling missing 'host_name' with 'unknown'
vancouver_df['host_name'] = vancouver_df['host_name'].fillna('Unknown')
london_df['host_name'] = london_df['host_name'].fillna('Unknown')
bangkok_df['host_name'] = bangkok_df['host_name'].fillna('Unknown')
```

# Standardizing Prices to CAD and Remove Duplicates

```python
# Convert London prices from GBP to CAD
london_df['price'] = london_df['price'] / 0.5958

# Convert Bangkok prices from THB to CAD
bangkok_df['price'] = bangkok_df['price'] / 26.7418

# Remove duplicates
vancouver_df.drop_duplicates(inplace=True)
london_df.drop_duplicates(inplace=True)
bangkok_df.drop_duplicates(inplace=True)
```
